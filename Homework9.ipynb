```python
import pyreadr
data = pyreadr.read_r('jperoaddata.RData')
data = data['jperoaddata']
data
```

```python
from statsmodels.formula.api import ols

model = ols('lndiffeall4mainancil ~ audit + und + fpm', data).fit(cov_type='HC1')
model.summary()
```

```python
model.pvalues.audit
```

```python
subsample = data[['lndiffeall4mainancil', 'audit', 'und', 'fpm', 'kecid']].dropna()
model = ols('lndiffeall4mainancil ~ audit + und + fpm', subsample).fit(cov_type='cluster', cov_kwds={'groups': subsample['kecid']})
model.summary()
```

```python
subsample = data[['lndiffeall4mainancil', 'auditstratnum', 'audit', 'und', 'fpm', 'kecid']].dropna()
model = ols('lndiffeall4mainancil ~ C(auditstratnum) + audit + und + fpm', subsample).fit(cov_type='cluster', cov_kwds={'groups': subsample['kecid']})
model.summary()
```

```python
model.bse
```

```python
model.pvalues
```
