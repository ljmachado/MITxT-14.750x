```python
data = pyreadr.read_r('AJRData.RData')
data = data['AJRData']
data
```

```python
subsample = data.query('baseco == 1.0')
subsample
```

```python
from statsmodels.formula.api import ols
model = ols('logpgp95 ~ avexpr', subsample).fit()
model.summary()
```
