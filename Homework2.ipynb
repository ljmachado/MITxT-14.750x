```python
from statsmodels.formula.api import ols
serious_data = data.query('seriousattempt == 1 and absnpolity2dummy11.notnull()')
serious_data['seriousattempt'] = pd.to_numeric(serious_data['seriousattempt'])
serious_data['absnpolity2dummy11'] = pd.to_numeric(serious_data['absnpolity2dummy11'])
serious_data['success'] = pd.to_numeric(serious_data['success'])
model = ols('absnpolity2dummy11 ~ success', serious_data).fit()
model.summary()
```

```python
model.pvalues
```

```python
model_fe = ols('absnpolity2dummy11 ~ success + weapondum2 + weapondum3 + weapondum4 + weapondum5 + weapondum6', serious_data).fit()
model_fe.summary()
```
