```python
import pyreadr
data = pyreadr.read_r('mitaData.RData')
data = data['mitaData']
data
```

```python
data['x2'] = data['x']**2
data['y2'] = data['y']**2
data['xy'] = data['x']*data['y']
data['x3'] = data['x']**3
data['y3'] = data['y']**3
data['x2y'] = data['x2']*data['y']
data['xy2'] = data['x']*data['y2']
data
```

```python
from statsmodels.formula.api import ols
subsample = data.query('d_bnd <= 100')
loc_poly = 'x	+ y	+ x2 + y2 + xy + x3 +	y3 + x2y + xy2'
ols = ols('lhhequiv ~ pothuan_mita + elv_sh + slope + infants + children + adults + bfe4_1 + bfe4_2 + bfe4_3 + '+loc_poly, subsample)
model = ols.fit(cov_type='cluster', cov_kwds={'groups': subsample['district']})
model.summary()
```
